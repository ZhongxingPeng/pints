```python
import sys
import os
cwd = os.getcwd()
sys.path.append('C://Users//bclamber//Documents//GitHub//pints')
import pints.plot # Use Pints' diagnostic tool
import numpy as np
import matplotlib.pyplot as plt
import time
```

# Few samples

Added an 'opacity' parameter that let's you control point transparency.

```python
n = 1000
samples = np.random.multivariate_normal([0,0,10], [[1,0.5,0.5],[0.5,1,-0.5],[0.5,-0.5,1]],n)
pints.plot.pairwise(samples, kde=False, opacity=1)
plt.show()
```

Changed the KDE part of the plots to make this more efficient and (I hope) more illuminating.

```python
t0 = time.time()
pints.plot.pairwise(samples, kde=True)

plt.show()
t1 = time.time()

print(t1 - t0)
```

# More points

```python
n = 10000
samples = np.random.multivariate_normal([0,0,10], [[1,0.5,0.5],[0.5,1,-0.5],[0.5,-0.5,1]],n)

t0 = time.time()
pints.plot.pairwise(samples, kde=True)

plt.show()
t1 = time.time()

print(t1 - t0)
```

# Many points - use thinning option to still allow plotting in reasonable time

```python
n = 100000
samples = np.random.multivariate_normal([0,0,10], [[1,0.5,0.5],[0.5,1,-0.5],[0.5,-0.5,1]],n)

t0 = time.time()
pints.plot.pairwise(samples, kde=True, thinning=10)

plt.show()
t1 = time.time()

print(t1 - t0)
```

# For fake data, add true values to plot

```python
n = 1000
samples = np.random.multivariate_normal([0, 5, 10], [[1, 0.5, 0.5],[0.5, 1, -0.5],[0.5, -0.5, 1]], n)
pints.plot.pairwise(samples, kde=True, true_values=[0, 5, 10])
plt.show()
```
